```python
import pandas as pd

# Load datasets
sales = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week5  customer_analysis/sales_data.csv")
customers = pd.read_csv("C:/Terraform/TRUPTI/The  developer arena/week5  customer_analysis/customer_churn.csv")

# Basic Information
print("Sales Shape:", sales.shape)
print("Customer Shape:", customers.shape)

# Check Missing Values
print(sales.isnull().sum())
print(customers.isnull().sum())

#Aggregation 1: Total Revenue
total_revenue = sales["Total_Sales"].sum()
print("Total Revenue =", total_revenue)

#Aggregation 2: Average Sales
avg_sales = sales["Total_Sales"].mean()
print("Average Sales =", avg_sales)

#Aggregation 3: Highest Sale
highest_sale = sales["Total_Sales"].max()
print("Highest Sale =", highest_sale)



#Group By Product
product_sales = sales.groupby("Product")["Total_Sales"].sum()
print(product_sales)



#pivot Table
pivot = pd.pivot_table(
    sales,
    values="Total_Sales",
    index="Region",
    columns="Product",
    aggfunc="sum"
)
print(pivot)



#Bar chart
import matplotlib.pyplot as plt
product_sales.plot(kind="bar")
plt.title("Sales by Product")
plt.show()


#Pie Chart
product_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.ylabel("")
plt.show()
```
